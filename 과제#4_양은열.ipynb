```python
#한글 설정

import matplotlib.font_manager

# 폰트 매니저
fm = matplotlib.font_manager.FontManager()

# 시스템에 등록된 폰트 목록을 가져옴
"""
font_names = [f.name for f in fm.ttflist]
print(font_names[:10])
"""
# 한글을 지원하는 폰트는 나오지 않음


from matplotlib import rcParams
rcParams["font.family"] = "AppleGothic"# 나눔고딕을 사용하려 했으나 맥에서는 나눔고딕이 없는것으로 추정. AppleGothic을 지원하는 것 같아 애플고딕 사용

rcParams["axes.unicode_minus"] = False

plt.xlabel("만족도")
plt.ylabel("가격")
plt.scatter([1,2,3], [4,5,6])
plt.title("손님 표본") # 애플고딕을 사용해서 한글 출력
```

```python
#여러 종류의 그래프 그리기

import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib import rcParams

sns.set()
%matplotlib inline

rcParams["font.family"] = "AppleGothic" #애플고딕
rcParams["axes.unicode_minus"] = False
```

```python
#선 그래프 그리는법

plt.plot([1,2,3], [4,1,2], label = "line1")
plt.plot([1,2,3], [5,6,0], label = "line2")
plt.legend(loc = "upper right") 
```

```python
#그래프 색 변경

plt.plot([1,2,3], [4,1,2], color = "purple", label = "line1") # 보라색
plt.plot([1,2,3], [5,6,0], color = "skyblue", label = "line2") #스카이 블루
plt.legend(loc = "upper right")
```

```python
plt.plot([1,2,3], [4,1,2], linestyle = "dashed", linewidth = 10, label = "line1") # 굵기가 10인 파선그래프
plt.plot([1,2,3], [5,6,0], linestyle = "dashed", linewidth = 1, label = "line2") # 굵기가 1인 파선그래프
plt.legend(loc = "upper right")
```

```python
#그래프 중간에 마커 삽입

plt.plot([1,2,3], [4,1,2], marker = "*", markersize = 5, label = "line1")
plt.plot([1,2,3], [5,6,0], marker = "*", markersize = 5, label = "line2") #크기 5짜리의 별모양 마커들 두개 line 1, line 2에 넣어주기
plt.legend(loc = "upper right")
```

```python
#퍼져있는 마커를 박은 그래프
#동그라미 모양 마커 1, 2 오렌지, 그레이

plt.scatter([1,2,3], [4,1,2], marker = "o", s = 80, color = "orange", label = "scatter1")
plt.scatter([1,2,3], [5,6,0], marker = "o", s = 80, color = "grey", label = "scatter2") 
plt.legend(loc = "upper right")
```

```python
#예시로 쓸 데이터

import pandas as pd
data = pd.DataFrame({"춤": [100, 80, 90], "노래": [100, 90, 80]}, 
                   index=["지효", "원영", "민주"])
display(data)
```

```python
sns.heatmap(data)
```

```python
sns.heatmap(data.values)
plt.xticks([0.5, 1.5], ["춤", "노래"])
plt.yticks([0.5, 1.5, 2.5], ["지효", "원영", "민주"]) 
```

```python
sns.heatmap(data, annot = True, fmt = "d")
```

```python
sns.heatmap(data, annot = True, fmt = ".2f")
```

```python
sns.heatmap(data, annot = True, fmt = ".2f", linewidth = 1, linecolor = "black")
```

```python
sns.heatmap(data, annot = True, fmt = ".2f", cmap = "Reds", linewidth = 1, linecolor = "Blue") #색깔
```
