```python
#merge
#on

df1 = pd.DataFrame({"직원": ["김철수", "박철수", "김민수", "이민수"],
                    "부서": ["회계", "데이터분석", "데이터분석", "인사"]})
df2 = pd.DataFrame({"직원": ["박철수", "이민수", "김민수", "김철수"],
                     "입사년도": [2012, 2016, 2020, 2024]})
merged_df = pd.merge(df1, df2, on = "직원")
display(merged_df) 
```

```python
#on

df3 = pd.DataFrame({"직원": ["김철수", "김민수", "김철수", "김민수"],
                   "부서": ["인사", "인사", "데이터분석", "데이터분석"],
                   "입사년도": [2012, 2016, 2020, 2024]})
df4 = pd.DataFrame({"직원": ["김철수", "김철수", "김민수", "김민수"],
                   "부서": ["인사", "데이터분석", "인사", "데이터분석"],
                   "월급": [500, 400, 350, 300]})
merged_df = pd.merge(df3, df4, on = ["직원", "부서"])
display(merged_df)
```

```python
#on

#left / right
df1.rename({"직원":"성함"}, axis = 1, inplace = True)
merged_data = pd.merge(df1, df2, left_on = ["성함"], right_on = ["직원"])
display(merged_data)
merged_data.drop("직원", axis = 1, inplace = True)
df3.rename({"직원":"성함", "부서":"부서/팀"}, axis = 1, inplace = True)
merged_data  = pd.merge(df3, df4, left_on = ["성함", "부서/팀"], right_on = ["직원", "부서"])
display(merged_data)
```

```python
# 좌 / 우 인덱스

df1 = pd.DataFrame({"직원": ["김철수", "박철수", "김민수", "이민수"], 
                    "부서": ["회계", "데이터분석", "데이터분석", "인사"]})
df2 = pd.DataFrame({"직원": ["박철수", "이민수", "김민수", "김철수"],
                    "입사년도": [2004, 2008, 2012, 2014]})
df1.set_index("직원", inplace = True)
display(df1)
merged_data = pd.merge(df1, df2, left_index = True, right_on = "직원")
display(merged_data)
display(df1.reset_index())
```

```python
#날짜 출력

a = pd.to_datetime("2020")
b = pd.to_datetime("2020-01")
c = pd.to_datetime("2020-01-03")
d = pd.to_datetime("2020-01-03 13")
e = pd.to_datetime("2020-01-03 13:30")
f = pd.to_datetime("2020-01-03 13:30:20")
print(a)
print(b)
print(c)
print(d)
print(e)
print(f) 

print(f.year) # 연도
print(f.month) # 월
print(f.day) # 일
print(f.hour) # 시
print(f.minute) # 분
print(f.second) # 초
```

```python
#연산

a = pd.to_datetime("2020-01-03")
b = pd.to_datetime("2020-01-05")
print(b - a)

five_day = pd.to_timedelta(5, "D") # 5일
five_hours = pd.to_timedelta(5, "h") # 5시간
five_minutes = pd.to_timedelta(5, "m") # 5분
five_seconds = pd.to_timedelta(5, "s") # 5초

print(a - five_day) # a에서 5일전
print(a + five_hours) # a에서 5시간 후
print(a + five_minutes) # a에서 5분 후
print(a - five_seconds) # a에서 5초 전
```
